```python
def weather(city_nameInput):
    # Python program to find current 
    # weather details of any city 
    # using openweathermap api 
    #https://openweathermap.org/current
    # import required modules 
    import requests, json 

    # Enter your API key here 
    api_key = "your own api key, if you would like one please ask"

    # base_url variable to store url 
    base_url = "http://api.openweathermap.org/data/2.5/weather?"

    # Give city name 
    city_name = city_nameInput

    # complete_url variable to store 
    # complete url address 
    complete_url = base_url + "appid=" + api_key + "&q=" + city_name 

    # get method of requests module 
    # return response object 
    response = requests.get(complete_url) 

    # json method of response object 
    # convert json format data into 
    # python format data 
    x = response.json() 

    # Now x contains list of nested dictionaries 
    # Check the value of "cod" key is equal to 
    # "404", means city is found otherwise, 
    # city is not found 
    if x["cod"] != "404": 

        # store the value of "main" 
        # key in variable y 
#Values of Main
#        "main": {
#    "temp": 282.55,
#    "feels_like": 281.86,
#    "temp_min": 280.37,
#    "temp_max": 284.26,
#    "pressure": 1023,
#    "humidity": 100
        y = x["main"]
    
        
        # store the value corresponding 
        # to the "temp" key of y 
        current_temperature = y["temp"] 
        current_temperature = current_temperature - 273.15 # changing Kelvin unit from to celcius degrees
        
        #getting the feels like value
        feels_like = y["feels_like"]
        feels_like = feels_like - 273.15# changing Kelvin unit from to celcius degrees
        
        #Storing the temp_max
        max_temp = y["temp_max"]
        max_temp = max_temp - 273.15#changing from kelvin to celcius degrees
        
        # getting the temp min
        min_temp = y["temp_min"] 
        min_temp = min_temp - 273.15#changing from kelvin to celcius degrees
        

        # store the value of "weather" 
        # key in variable z 
        z = x["weather"] 
        
        # store the value corresponding 
        # to the "description" key at 
        # the 0th index of z 
        weather_description = z[0]["description"] 
        
        #Getting the wind speed
        wind = x["wind"]
        wind_speed = wind["speed"]
        wind_speed =  wind_speed * 3.6 # changing from meter/sec to Km/hr

        # print following values 
        print("City Name : " + city_name)
        print(str(weather_description) +
                "\nTemperature (in Celcius Degree's) : " + str(round(current_temperature,2)) +
                "\nFeels like : "+ str(round(feels_like,2))+
                "\nMax Temp :" + str(round(max_temp,2))+
                "\nMin Temp :" + str(round(min_temp,2))+
                "\nWind Speed : "+str(round(wind_speed,2))+" km/hr"
             ) 

    else: 
        print(" City Not Found ") 
    
```
